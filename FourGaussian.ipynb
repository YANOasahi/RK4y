Import some libraries<br>
Note that plt is just for print the output of FourGaussian

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Definition of FourGaussian<br>
This is just the sum of four gaussians. I assumed that all gaussians have the same amplitude and sigma.

```python
def FourGaussian(x,amp,sigma,mean1,mean2,mean3,mean4):
    return amp*np.exp(-(x-mean1)**2/(sigma**2))+amp*np.exp(-(x-mean2)**2/(sigma**2))+amp*np.exp(-(x-mean3)**2/(sigma**2))+amp*np.exp(-(x-mean4)**2/(sigma**2))
```

To check the shape of FourGaussian, plot it

```python
x=np.arange(-50,150,0.5)


plt.plot(x,FourGaussian(x,100,20,15,35,55,75))
```
