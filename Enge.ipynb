Import some libraries.<br>
Note that plt is just for checking the shape of this function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Definition of Enge function

```python
def Enge(x,first,second,third,fourth):
    return 1/(1+(np.exp(first+(x*second)+(x**2*third)+(x**3*fourth))))
```

To check the shape of Enge, plot it.

```python
x=np.arange(-10,10,0.5)

plt.plot(x,Enge(x,0.95,1,0.2,0.05))
```
