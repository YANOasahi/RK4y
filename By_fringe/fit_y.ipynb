```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


data = np.genfromtxt('./y_by_x600_z0.dat')

def fifth(x,a,b,c,d):
    return a*np.power(x,3)+b*np.square(x)+c*x+d

ydata = np.array(data[:, 1])
bydata = np.array(data[:, 3])

# フィッティング
# 初期値の設定
p0 = [1,10,10,0.1]

# curve_fitを用いて最適化
popt, pcov = curve_fit(fifth, ydata, bydata, p0=p0, maxfev=5000)

# カイ二乗の計算
y_fit_data = fifth(ydata, *popt)  # データポイントでのフィット値
residuals = ydata - y_fit_data            # 残差
chi_squared = np.sum((residuals)**2)     # カイ二乗値 (誤差を1と仮定)
reduced_chi_squared = chi_squared / (len(ydata) - len(popt))  # 自由度で割る

# 最適化結果の表示
a_fitted, b_fitted, c_fitted, d_fitted = popt
print("Fitted parameters:")
print(f'Chi-squared: {chi_squared}')
print(f'reduced chi-squared: {reduced_chi_squared}')

# プロット用のデータ生成
y_fitted = np.linspace(min(ydata), max(ydata), 1000)
by_fitted = fifth(y_fitted, *popt)

# プロット
plt.figure(figsize=(8, 6))
plt.scatter(ydata, bydata, label="Data", color="blue")
plt.plot(y_fitted, by_fitted, label="fit", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("fifth order function")
plt.show()
```
